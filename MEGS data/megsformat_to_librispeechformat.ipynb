```python
import os
from shutil import copyfile
import codecs
from tqdm import tqdm

#Open files.txt and go through it line by line 
filestxt = codecs.open('files.txt', 'r', 'utf-8') 
filesLines = filestxt.readlines()
total_lines = len(filesLines)

#Open labels_word-transcript.txt
labeltxt = codecs.open('labels_word-transcript.txt', 'r', 'utf-8')
labelLines = labeltxt.readlines()
labelIds = []
for labelline in labelLines:
    labelIds.append(labelline.split(" 0.0 -1 ")[0])
print("Found ", len(labelIds), "labelIds")

traindir = 'librispeechformat/data/train'
if not os.path.exists(traindir):
    os.makedirs(traindir)
filecounter = 0
fileslinecount = 0
currentfolder2path = ""
folderswapped = False
labelsfilestring = ""
linesperfile = 0
for fileline in filesLines:
    fileslinecount += 1
    #Extract infos
    id = fileline.split(' ')[0]
    file_path = fileline.split(' ')[1].rstrip()
    #filename = file_path.split('/')[-1]
    label = labelLines[labelIds.index(id)].split(' -1 ')[-1].rstrip().upper()

    #print(id)
    #print("file_path:" + file_path + "|||||")
    #print(filename)
    #print(label)
    
    #create folders and subfolders
    filesPerDestFolder1 = 100000
    filesPerDestFolder2 = 25000
    folder1name = str(int((fileslinecount - 1)  / filesPerDestFolder1))
    folder2name = str(int((fileslinecount  - 1) / filesPerDestFolder2))
    folder1path = traindir + '/' + folder1name
    folder2path = traindir + '/' + folder1name + '/' + folder2name
    
    if currentfolder2path != folder2path:
        folderswapped = True
        currentfolder2path = folder2path
        print("FILESTART")
        linesperfile = 0
        filecounter += 1
        print("File Nummer: ", filecounter)
    
    if not os.path.exists(folder1path):
        os.mkdir(folder1path)
    if not os.path.exists(folder2path):
        os.mkdir(folder2path)
    
    #copy audio file, whereas folder names and id are part of filename, so it is unique
    newfilename = folder1name + '-' + folder2name + '-' + id
    #print(newfilename)
    dest = folder2path + "/" + newfilename
    #print(dest)
    #copyfile(file_path, dest)
    
    #copy labels
    if folderswapped == True:
        labelsfilestring = ""
    
    linesperfile +=1
    labelsfilestring += newfilename + " " + label + "\n"
    #print(labelsfilestring[0:1000])
    
    #last item of file
    if fileslinecount % filesPerDestFolder2 == 0 or fileslinecount == total_lines:
        print("linesperfile: ", linesperfile)
        print("FILEEND")
        curr_label_filepath = folder2path + "/" + folder1name + '-' + folder2name + '.' + 'trans.txt'
        with codecs.open(curr_label_filepath, "w", 'utf-8') as curr_label_file:
          curr_label_file.write(labelsfilestring.rstrip())

            
    #reset variables
    folderswapped = False
```
